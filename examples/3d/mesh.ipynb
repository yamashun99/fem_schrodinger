```python
import gmsh

# Gmshの初期化
gmsh.initialize()

# 新しいモデルを作成
gmsh.model.add("simple_cube")

# 立方体を作成（中心座標：(0.5, 0.5, 0.5)、サイズ：1）
gmsh.model.occ.addBox(0, 0, 0, 1, 1, 1)

# ジオメトリの同期
gmsh.model.occ.synchronize()

# メッシュ生成
gmsh.model.mesh.generate(3)

# GUIで表示
gmsh.fltk.run()

# メッシュを保存
gmsh.write("simple_cube.msh")

# Gmshの終了
gmsh.finalize()
```

```python
import gmsh

# Gmshの初期化
gmsh.initialize()
gmsh.model.add("model")

# モデル生成などの処理をここに記述...

# 立方体を作成（中心座標：(0.5, 0.5, 0.5)、サイズ：1）
gmsh.model.occ.addBox(0, 0, 0, 1, 1, 1)

# ジオメトリの同期
gmsh.model.occ.synchronize()

# メッシュ生成
gmsh.model.mesh.generate(3)

# 頂点の座標と番号の取得
node_ids, node_coords, _ = gmsh.model.mesh.getNodes()
print("頂点の座標と番号:")
for i, coord in zip(node_ids, node_coords):
    print(f"頂点 {i}: {coord}")

# 要素の取得
element_types, element_ids, element_node_ids = gmsh.model.mesh.getElements()
print("\n要素とその頂点:")
for elem_type, elem_ids, node_ids in zip(element_types, element_ids, element_node_ids):
    for elem_id, nodes in zip(elem_ids, node_ids):
        print(f"要素 {elem_id} (タイプ {elem_type}): 頂点 {nodes}")

# 境界の取得（例として物理グループ1のエンティティを取得）
boundary_nodes = gmsh.model.mesh.getNodesForPhysicalGroup(2, 1)
print("\n境界の頂点番号:")
print(boundary_nodes)

# メッシュを保存
gmsh.write("simple_cube.msh")

# Gmshの終了
gmsh.finalize()
```

```python
import gmsh

gmsh.initialize()
gmsh.model.add("model")

# ここにモデル生成のコードを追加
# 立方体を作成（中心座標：(0.5, 0.5, 0.5)、サイズ：1）
gmsh.model.occ.addBox(0, 0, 0, 1, 1, 1)

# ジオメトリの同期
gmsh.model.occ.synchronize()

# メッシュ生成
gmsh.model.mesh.generate(3)

# 頂点座標と番号の取得
nodes = gmsh.model.mesh.getNodes()
print("頂点の座標と番号:")
for i in range(0, len(nodes[0])):
    print(f"頂点番号 {nodes[0][i]}: 座標 {nodes[1][i*3:i*3+3]}")

# 要素とその頂点番号の取得
elements = gmsh.model.mesh.getElements()
element_types, element_tags, node_tags = elements
for i, e_type in enumerate(element_types):
    print(f"\n要素タイプ {e_type}:")
    for j, e_tag in enumerate(element_tags[i]):
        print(f"  要素番号 {e_tag}: 頂点番号 {node_tags[i][j]}")

# 物理グループの頂点番号を取得（例: 物理グループ1）
group_nodes = gmsh.model.mesh.getNodesForPhysicalGroup(2, 1)
print("\n境界の頂点番号:")
print(group_nodes[0])

gmsh.finalize()
```
