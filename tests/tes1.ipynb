```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_A(N):
    h = 1/(N)
    A = np.zeros((N, N))
    for i in range(N):
        if i == 0:
            A[i, i] = -2
            A[i, i+1] = 1
        elif i == N-1:
            A[i, i] = -2
            A[i, i-1] = 1
        else:
            A[i, i] = -2
            A[i, i-1] = 1
            A[i, i+1] = 1
    A[0, 0] = -1
    A = -A/h**2
    return A
```

```python
N = 3
A = make_A(N)
f = np.ones(N)
f[0] = 0.5
u = np.linalg.solve(A, f)
x = np.linspace(0, 1, N+1)
u_new1 = np.append(u, 0)
plt.plot(x, u_new1)
plt.plot(x, -x**2/2 + 1/2)
```

```python
errs = []
for N in range(10, 1000, 10):
    A = make_A(N)
    f = np.ones(N)
    f[0] = 0.5
    u = np.linalg.solve(A, f)
    x = np.linspace(0, 1, N+1)
    #u_new0 = np.insert(u, 0, u[0])
    u_new1 = np.append(u, 0)
    plt.plot(x, u_new1)
    errs.append(0.5-u[0])
```

```python
fig, ax = plt.subplots()
ax.plot(1/np.arange(10, 1000, 10), errs)
```
